```python
from mp_api.client import MPRester

sparks_API = "GZk6ELwctMt7Mskij1TEPw7nEFIxtzEr"

with MPRester(sparks_API) as mpr:
    # Searching for materials that contain Oxygen and at least one other element
    results = mpr.materials.summary.search(
        elements=["O"],  # Ensures oxygen is present
        num_elements=(2, 10),  # Ensures it's a compound, not pure O3
        fields=["formula_pretty", "material_id"]  # ✅ Use 'formula_pretty' instead of 'pretty_formula'
    )

    # Filter only those containing "O3" explicitly in the formula
    filtered_results = [m for m in results if "O3" in m.formula_pretty]

    # Print results
    if filtered_results:
        print(f"Found {len(filtered_results)} materials containing 'O3':")
        for material in filtered_results:
            print(f"Material ID: {material.material_id}, Formula: {material.formula_pretty}")
    else:
        print("No results found.")
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
iris.data
```

```python
iris.target
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df
```

```python
x = df.iloc[:,:-1]
y= iris.target
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
##by default post pruning
treemodel = DecisionTreeClassifier()
```

```python
treemodel.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
#hyperparameter tuning
treemodelu = DecisionTreeClassifier(max_depth=2)
## bcz in depth two 34, 30 has the max sample
```

```python
treemodelu.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodelu,filled=True)
```

```python
y_pred = treemodelu.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report

```

```python
score = accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
